```python
def reverse_Matrix(matrix):
  row,col = matrix.shape

  new_matrix = np.zeros((row,col),dtype = int)
  for i in range(row):
    for j in range(col):
      new_matrix[i][j]=matrix[row-i-1][col-j-1]

  return new_matrix


matrix = np.array([
[14,  8,  0,  4],
[9,  8,  13,  13],
[9,  3,  1,  4],
[2,  10,  13,  6]
])
print_matrix(matrix)
print('################')
print()
reversed_matrix = reverse_Matrix(matrix)
print_matrix(reversed_matrix)

#This should print
#|  6  |  13 |  10 |  2  |
#-------------------------
#|  4  |  1  |  3  |  9  |
#-------------------------
#|  13  |  13  |  8 |  9 |
#-------------------------
#|  4 |  0  |  8  |  14  |
#-------------------------
```

Task 5: Matrix Compression

```python
def compress_matrix(mat):
    #TO DO
    row,col = mat.shape
    new_arr = np.zeros((row//2,col//2),dtype = int)
    m = 0
    n = 0
    for i in range(0,row,2):
      for j in range(0,col,2):
        sum = 0
        sum += mat[i][j]+mat[i+1][j]+mat[i+1][j+1]+mat[i][j+1]
        # print(sum)
        new_arr[m][n]=sum
        n+=1
      m+=1
      n=0
    return new_arr



matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]
                 ])
print_matrix(matrix)
print('################')
print()
returned_array=compress_matrix(matrix)
print_matrix(returned_array)
#This should print
#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```

Task 6: Game Arena

```python
def play_game(arena):
  #TO DO
  # row,col = arena.shape
  count = 0
  points = 0
  #print(len(arena))
  for i in range(len(arena)):
      for j in range(len(arena[i])):
          if arena[i][j] % 50 == 0 and arena[i][j]!=0:
              count += 1
              if i > 0 and arena[i-1][j] == 2:
                  points += 2
                  #print("up",points)
              if i < len(arena)-1 and arena[i+1][j] == 2:
                  points += 2
                  #print("down",points)
              if j > 0 and arena[i][j-1] == 2:
                  points += 2
                  #print("left",points)
              if j < len(arena[i])-1 and arena[i][j+1] == 2:
                  points += 2
                  #print("right",points)
  if points >= 10:
      print(f"Points Gained: {points}. Your team has survived.")
  else:
      print(f"Points Gained: {points}. Your team is out.")

arena=np.array([[0,2,2,0],
                [50,1,2,0],
                [2,2,2,0],
                [1,100,2,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 6. Your team is out.

print(".....................")
arena=np.array([[0,2,2,0,2],
                [1,50,2,1,100],
                [2,2,2,0,2],
                [0,200,2,0,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 14. Your team has survived the game.
```
