```python
#Assignment Part 1
class Patient:#node
  #write a constructor
  def __init__(self,id,name,age,blood_group):
    self.id = id
    self.name = name
    self.age = age
    self.blood_group = blood_group
```

```python
import numpy as np

class Node:
  def __init__(self, elem, next = None, prev = None):
    self.elem = elem
    self.next = next
    self.prev = prev
```

```python
class WRM:
  #write a constructor
  def __init__(self):
    self.head = Node(None)
    self.tail = self.head
    self.size = 0

  def registerPatient(self,id, name, age, bloodgroup):
    #To Do
    new_patient = Patient(id, name, age, bloodgroup)
    new_node = Node(new_patient)

    if self.size == 0:
      self.head.next = new_node
      new_node.prev = self.head
    else:
      new_node.prev = self.tail
      self.tail.next = new_node

    new_node.next = self.head
    self.head.prev = new_node

    self.tail = new_node
    self.size += 1

  def servePatient(self):
    #To Do
    if self.size == 0:
        print("No patients in the waiting room.")
    else:
        first_patient = self.head.next
        self.head.next = first_patient.next
        first_patient.next.prev = self.head
        self.size -= 1
        print("Serving patient:", first_patient.elem.name)

  def showAllPatient(self):
    #To Do
    if self.size == 0:
      print("No patients in the waiting room.")
    else:
      current_node = self.head.next
      while current_node != self.head:
        print("Patient ID:", current_node.elem.id)
        current_node = current_node.next

  def canDoctorGoHome(self):
    #To Do
    return self.size == 0


  def cancelAll(self):
    #To Do
    if self.size == 0:
      print("Not")
    else:
      self.head.next = None
      self.tail = self.head
      self.size = 0

  def ReverseTheLine(self):
    #To Do
    if self.size == 0:
      print("No patients to reverse")
    else:
      current_node = self.tail
      while current_node != self.head:
        print("Patient ID:", current_node.elem.id)
        current_node = current_node.prev


```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    e.next = None
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
  #TO DO
  count = 0
  for i in string:
    if i == "<":
      stack.push(i)
    elif i == ">":
      if stack.peek() == "<":
        stack.pop()
        count+=1
  return count



print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```
